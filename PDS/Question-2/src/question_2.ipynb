```python
# prompt: read csv

import pandas as pd

# Assuming your CSV file is in the same directory as your notebook
# Replace 'your_file.csv' with the actual name of your CSV file
df = pd.read_csv('https://raw.githubusercontent.com/Likhitha02-tech/PDS/refs/heads/main/StudentsPerformance.csv')

# Print the first 5 rows of the DataFrame
print(df.head())
```

```python
# prompt:  Check for non-numeric values in the 'math score

import pandas as pd
# Check for non-numeric values in the 'math score' column
non_numeric_rows = df[pd.to_numeric(df['math score'], errors='coerce').isnull()]

if not non_numeric_rows.empty:
  print("Non-numeric values found in 'math score' column:")
  print(non_numeric_rows)
else:
  print("No non-numeric values found in 'math score' column.")
```

```python
# prompt: Gender-wise Average Scores Comparison

# Calculate the average math score for each gender
average_scores_by_gender = df.groupby('gender')['math score'].mean()
# Print the results
print("Average Math Scores by Gender:\n", average_scores_by_gender)
```

```python
# prompt: Distribution of Scores in Math, Reading, and Writing

import matplotlib.pyplot as plt
import seaborn as sns

# Create histograms for math, reading, and writing scores
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
sns.histplot(df['math score'], kde=True)
plt.title('Distribution of Math Scores')

plt.subplot(1, 3, 2)
sns.histplot(df['reading score'], kde=True)
plt.title('Distribution of Reading Scores')

plt.subplot(1, 3, 3)
sns.histplot(df['writing score'], kde=True)
plt.title('Distribution of Writing Scores')

plt.tight_layout()
plt.show()
```

```python
# prompt: print conclusion

print("Conclusion:")
print("Based on the analysis of the student performance dataset:")
print("- Male and female students have different average math scores.")
print("- The distributions of math, reading, and writing scores provide insights into the spread of student performance in these areas.")
print("- Further analysis can be conducted to explore correlations between variables, identify factors influencing student performance, and understand potential disparities in educational outcomes.")
```

```python
# prompt: Bar Chart: Gender vs Average Math Score

import matplotlib.pyplot as plt
plt.figure(figsize=(8, 6))
plt.bar(average_scores_by_gender.index, average_scores_by_gender.values)
plt.xlabel('Gender')
plt.ylabel('Average Math Score')
plt.title('Average Math Score by Gender')
plt.show()
```

```python
# prompt: print conclusion

print("Conclusion:")
print("Based on the analysis of the student performance dataset:")
print("- Male and female students have different average math scores.")
print("- The distributions of math, reading, and writing scores provide insights into the spread of student performance in these areas.")
print("- Further analysis can be conducted to explore correlations between variables, identify factors influencing student performance, and understand potential disparities in educational outcomes.")
```

```python
# prompt: # Prepare the data by calculating the mean scores for each subject by lunch type

# Calculate the mean scores for each subject by lunch type
mean_scores_by_lunch = df.groupby('lunch')[['math score', 'reading score', 'writing score']].mean()

# Print the results
print("Mean Scores by Lunch Type:\n", mean_scores_by_lunch)
```

```python
# prompt:  Box Plot: Parental Education Level vs Distribution of Reading Scores

import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
sns.boxplot(x='parental level of education', y='reading score', data=df)
plt.title('Parental Education Level vs. Reading Score Distribution')
plt.xlabel('Parental Level of Education')
plt.ylabel('Reading Score')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
# prompt: print conclusion

print("Conclusion:")
print("Based on the analysis of the student performance dataset:")
print("- Students with standard lunch tend to perform better in math, reading, and writing compared to those with free/reduced lunch.")
print("- Students whose parents have a master's degree tend to have higher reading scores compared to students with parents who have a high school education.")
print("- Male students tend to have slightly higher average math scores than female students.")
print("- Further analysis can be conducted to explore correlations between variables, identify factors influencing student performance, and understand potential disparities in educational outcomes.")
```

```python
# prompt: bar Plot: Math Score vs Reading Score vs race/ethnicity

import matplotlib.pyplot as plt

# Group the data by race/ethnicity and calculate the mean math and reading scores
mean_scores_by_race = df.groupby('race/ethnicity')[['math score', 'reading score']].mean()

# Create a bar plot
ax = mean_scores_by_race.plot(kind='bar', figsize=(10, 6))

# Set the title and labels
plt.title('Average Math and Reading Scores by Race/Ethnicity')
plt.xlabel('Race/Ethnicity')
plt.ylabel('Average Score')

# Add a legend
plt.legend(['Math Score', 'Reading Score'])

# Rotate x-axis labels for better readability
plt.xticks(rotation=45, ha='right')

# Show the plot
plt.show()
```

```python
# prompt: print conclusion for above plot

print("Conclusion:")
print("Based on the analysis of the student performance dataset:")
print("- Group E students have the highest average math and reading scores, while group A students have the lowest.")
print("- There is a noticeable difference in average scores across different race/ethnicity groups.")
print("- Further analysis can be conducted to investigate the potential socio-economic factors and educational resources that contribute to these disparities.")
```

```python
# prompt: Histogram: Distribution of Math Scores

import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8, 6))
sns.histplot(df['math score'], kde=True)
plt.title('Distribution of Math Scores')
plt.xlabel('Math Score')
plt.ylabel('Frequency')
plt.show()
```

The histogram shows a **normal distribution** of math scores, peaking around **60-70**. Most students scored in this range, with fewer scoring extremely low or high. The smooth line confirms the bell-shaped pattern.

```python
# prompt:  Heatmap: Correlation Between Math, Reading, and Writing Scores

import matplotlib.pyplot as plt
plt.figure(figsize=(8, 6))
sns.heatmap(df[['math score', 'reading score', 'writing score']].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Between Math, Reading, and Writing Scores')
plt.show()
```

The heatmap shows strong positive correlations between math, reading, and writing scores. Key points:

- **Math vs. Reading**: Correlation of **0.82**, indicating that good math performers tend to do well in reading.
- **Reading vs. Writing**: Very high correlation of **0.95**, showing a strong link between these two.
- **Math vs. Writing**: Correlation of **0.80**, meaning there's also a strong connection between math and writing performance.

The redder the color, the stronger the correlation.

```python
# prompt:  Line Plot: Average Scores Across Subjects by Ethnicity Group

import matplotlib.pyplot as plt
# Calculate the average scores for each subject by ethnicity group
average_scores_by_ethnicity = df.groupby('race/ethnicity')[['math score', 'reading score', 'writing score']].mean()

# Create a line plot
plt.figure(figsize=(10, 6))
plt.plot(average_scores_by_ethnicity.index, average_scores_by_ethnicity['math score'], label='Math Score')
plt.plot(average_scores_by_ethnicity.index, average_scores_by_ethnicity['reading score'], label='Reading Score')
plt.plot(average_scores_by_ethnicity.index, average_scores_by_ethnicity['writing score'], label='Writing Score')

plt.xlabel('Ethnicity Group')
plt.ylabel('Average Score')
plt.title('Average Scores Across Subjects by Ethnicity Group')
plt.xticks(rotation=45, ha='right')
plt.legend()
plt.grid(True)
plt.show()
```
